```python
import numpy as np
import pandas as pd

data = pd.read_json('data.json')
# df = pd.DataFrame(data)

# data
```

```python
# We'll access restaurants and do a function on each object and replace it
df = pd.DataFrame(data)
ratings_array = []
def expenses(val):
    print(val)
    obj = val
    # obj = obj['restaurants']
    cost = 0
    cnt = 0
    if(type(obj) == float):
        # NaN value preprocessed
        return 0
    
    if('rating' in obj):
        lc_obj = {"name" : obj['name'], "rating": 0 , "address" : "Not in data"}
        try:
            lc_obj['rating'] = float(obj['rating'])
            cnt = cnt+1
            ratings_array.append(lc_obj)
            
        except:
            lc_obj = lc_obj
        
        try:
            lc_obj['address'] = obj['address']
        except:
            lc_obj = lc_obj
        
        return lc_obj
    else:
       return np.nan
        
    
# 12 problem 
df2 = df
df2 = df2['Bangalore']
df2 = df2.iloc[0]
df2 = pd.DataFrame(df2)
# print(df2)
df2 = df2['restaurants'].apply(expenses)
```

```python
# ratings_array
ratings_df = pd.DataFrame(ratings_array)
print(ratings_df)
ratings_df = ratings_df.drop(columns=['address'])
ratings_df
```
