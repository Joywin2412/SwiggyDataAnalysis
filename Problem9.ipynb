```python
# We'll access restaurants and do a function on each object and replace it
df = pd.DataFrame(data)
def expenses(val):
    # print(val)
    obj = val
    obj = obj['restaurants']
    # print(obj)
    cost = 10000000000
    cnt = 0
    
    
    for restaurant in obj:
        # print(restaurant)
        
        if('cost' in obj[restaurant]):
            if(type(obj[restaurant]['cost']) == list):
                # print(obj[restaurant]['cost'])
                try:
                    cost = min(cost,int(obj[restaurant]['cost'][-1]))

                    # print(int(obj[restaurant]['cost'][-1]))
                    cnt = cnt+1
                except:
                    continue
            else:
                obj[restaurant]['cost'] = obj[restaurant]['cost'].split()
                try:
                    cost = min(cost,float(obj[restaurant]['cost'][-1]))
                    cnt = cnt+1
                except:
                    continue
                    # print(obj[restaurant]['cost'][-1])
        elif('price' in obj[restaurant]):
           
            print(obj[restaurant])
            str = obj[restaurant]['price'].split()
            str = str[0]
            str = str.replace("â‚¹","")
            cost  = min(cost,float(str))
            cnt = cnt+1
        

    # print(cost)

    if(cost ==10000000000 ) :
        return np.nan
    else:
        return cost
# data
# df2 = df2.drop('link')
df2 = df
df2 = df2.drop('link')
df2 = df2['Delhi']
df2 = df2.dropna()
df2 = pd.DataFrame(df2)
# print(df2)
# for i in range(0,len(df2.index)):
#     df2 = df.iloc[i].apply(expenses)
df3 = df2['Delhi'].apply(expenses)
```
